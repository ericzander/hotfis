## Setup  <a name="c0"></a>

```python
import hotfis as hf

# Import matplotlib for plotting functionality
import matplotlib.pyplot as plt
```

---

## Takagi-Sugeno-Kang ('tsk')

```python
fn = hf.FuzzyFunc("hot", [0], "tsk")

plt.figure(figsize=(4, 2.5), dpi=300)
fn.plot(-2, 2)
#plt.savefig("img/fuzzyfuncs/tsk.png", facecolor="white", bbox_inches="tight")
plt.show()
```

## Triangular ('triangular')

```python
fn = hf.FuzzyFunc("hot", [-1, 0, 1], "triangular")

plt.figure(figsize=(4, 2.5), dpi=300)
fn.plot(-2, 2)
# plt.savefig("img/fuzzyfuncs/triangular.png", facecolor="white", bbox_inches="tight")
plt.show()
```

## Trapezoidal ('trapezoidal')

```python
fn = hf.FuzzyFunc("hot", [-2, -1, 1, 2], "trapezoidal")

plt.figure(figsize=(4, 2.5), dpi=300)
fn.plot(-3, 3)
# plt.savefig("img/fuzzyfuncs/trapezoidal.png", facecolor="white", bbox_inches="tight")
plt.show()
```

## Left Edge ('leftedge')

```python
fn = hf.FuzzyFunc("hot", [-1, 1], "leftedge")

plt.figure(figsize=(4, 2.5), dpi=300)
fn.plot(-2, 2)
# plt.savefig("img/fuzzyfuncs/leftedge.png", facecolor="white", bbox_inches="tight")
plt.show()
```

## Right Edge ('rightedge')

```python
fn = hf.FuzzyFunc("hot", [-1, 1], "rightedge")

plt.figure(figsize=(4, 2.5), dpi=300)
fn.plot(-2, 2)
# plt.savefig("img/fuzzyfuncs/rightedge.png", facecolor="white", bbox_inches="tight")
plt.show()
```

## Gaussian ('gaussian')

> Params: [mean, standard deviation]

```python
# Create standard normal
fn = hf.FuzzyFunc("hot", [0, 1], "gaussian")

plt.figure(figsize=(4, 2.5), dpi=300)
fn.plot(-3, 3)
# plt.savefig("img/fuzzyfuncs/gaussian.png", facecolor="white", bbox_inches="tight")
plt.show()
```
